```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('online_shoppers_intention.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
for i in df.columns:
    if df[i].dtype == 'object':
        print(df[i].value_counts())
```

```python
visitors = pd.get_dummies(df['VisitorType'])
x = pd.concat([df,visitors],axis=1)
x.head()
```

```python
visitors
```

```python
df.info()
```

```python
y = x['Revenue']
y
```

```python
x_new = x.drop(['Revenue','Month','VisitorType'],axis=1)
```

```python
x_new['Weekend'] = np.asarray(x_new['Weekend'].astype(np.float32))
y = np.asarray(y).astype(np.float32)
```

```python
x_new.info()
```

```python
y.shape
```

```python
x_new.shape
```

```python
model = Sequential()
model.add(Dense(18,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))

model.add(Dense(units=2,activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_new,y,epochs=100,validation_split=0.1)
```
